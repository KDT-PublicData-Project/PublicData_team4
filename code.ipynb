```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
# 지역별 전체 평균 월급 및 임금상승률
salaryDF = pd.read_excel('../publicdataProject/data/월평균_임금_및_임금상승률_시도__20240129151046.xlsx',index_col=0,header=[0,1])
salaryDF
```

```python
# 데이터 표준화 및 결측치 제거

salaryDF.replace('-',np.nan, inplace=True)
salaryDF.rename(index = {'강원특별자치도':'강원도'}, inplace=True)
salaryDF
```

```python
# 지역별 임금상승률 뽑은 DF
salaryflowDF = salaryDF.xs('임금상승률 (%)', level=1, axis=1)
salaryflowDF.dropna(inplace=True)
salaryflowAllDF = salaryflowDF.drop(labels = '전국')

# 지역별 평균 임금 상승률 컬럼 추가
meanList = []
for i in salaryflowAllDF.index:
    meanList.append(salaryflowAllDF.loc[i].mean())
salaryflowAllDF['평균 임금상승률'] = meanList
salaryflowAllDF.sort_values(by='평균 임금상승률', ascending= False,inplace=True)

# 년도별 평균 임금 상승률 행 추가
salaryflowAllDF
yearList = []
for i in salaryflowAllDF.columns[:-1]:
    yearList.append(salaryflowAllDF[i].mean())
yearList.append('-')
yearList
salaryflowAllDF.loc['연도별 평균 임금상승률'] = yearList

# 전국과 대구광역시 임금상승률 비교
salaryflowDagDF = pd.concat([salaryflowDF.loc['전국'],salaryflowDF.loc['대구광역시']], axis=1)
salaryflowDagDF
```

```python
w = 0.3
nrow = salaryflowDagDF.shape[0] # 행의 갯수
idx = np.arange(nrow) #행의 갯수를 리스트로

plt.figure(figsize = (12, 6))
plt.bar(idx - w/2, salaryflowDagDF['대구광역시'],color = 'red', width = w)
plt.title('임금 상승률')
plt.bar(idx + w/2, salaryflowDagDF['전국'], color = 'gray', width = w)
plt.legend(['대구광역시','전국'])
plt.xticks(idx, salaryflowDagDF.index)
plt.show()
```

```python
# 전국을 제외한 지역별 전체 평균 월급(2023년)
salary2023DF = salaryDF.drop(labels='전국')
salary2023DF = salary2023DF.xs('상용 월평균 임금 (원)',level=1, axis=1)

salary2023DF = salary2023DF['2023'] /10000
salary2023DF.sort_values(inplace=True)
salary2023DF
```

```python
plt.figure(figsize=(12,6))
plt.barh(salary2023DF.index,salary2023DF,color = 'gray')
plt.barh('대구광역시',salary2023DF['대구광역시'],color = 'red')
plt.title('[2023년도 지역별 평균 월급]')
plt.xlabel('평균 월급 (만원)')
plt.xlim(300)
plt.axvline(salary2023DF['대구광역시'], color = 'red', alpha = 0.5)
plt.axvline(salary2023DF['세종특별자치시'], color = 'k', alpha = 0.5)
plt.text(y= '세종특별자치시', x = 365, s= '세종특별자치시 평균',size = 13,  ha = 'left', va = 'center')
plt.text(y= '대구광역시', x = 334, s= '대구광역시 평균',size = 13,  ha = 'left', color = 'red', va = 'center')
```

```python
aiDF = pd.read_excel('../publicdataProject/data/행정구역_시도___산업_규모별_임금_및_근로시간_상용근로자_상용근로자_5인이상_사업체__20240129154755.xlsx',index_col=0, header = [0,1])
aiDF.drop(labels = [('산업분류','산업분류'),('규모별','규모별')], axis = 1, inplace=True)
aiDF.rename(index = {'전국(5인이상)':'전국','경기':'경기도','서울':'서울특별시','전남':'전라남도','부산':'부산광역시','광주':'광주광역시','대전':'대전광역시','충남':'충청남도','강원':'강원도',
'대구':'대구광역시','인천':'인천광역시','제주':'제주특별자치도','경북':'경상북도','충북':'충청북도','경남':'경상남도','전북':'전라북도','울산':'울산광역시','세종':'세종특별자치시'},inplace = True)
aiDF.drop(labels='전국',inplace=True)
aigenDF = aiDF.xs('상용월급여액 (원)', level= 1, axis=1).T

aigenperDF = aigenDF.pct_change()*100
aigenperDF.dropna(inplace=True)
aigen2021DF = aigenperDF.loc['2021']
aigen2022DF = aigenperDF.loc['2022']
aigen2023DF = aigenperDF.loc['2023']
aigenDF2 = pd.concat([aigenDF.loc['2020'], aigenDF.loc['2023']], axis = 1).T
aigenDF2 = aigenDF2.pct_change()*100
aigenDF2.dropna(inplace=True)
aigengggDF2 = pd.concat([aigenDF2.서울특별시, aigenDF2.부산광역시, aigenDF2.인천광역시, aigenDF2.광주광역시,aigenDF2.울산광역시,aigenDF2.대구광역시, aigenDF2.대전광역시]) 
ggname = ['서울특별시','부산광역시','인천광역시','광주광역시','울산광역시','대구광역시','대전광역시']
aigengggDF2.index = ggname
aigengggDF2.sort_values(ascending=False, inplace=True)
```
