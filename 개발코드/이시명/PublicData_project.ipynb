```python
city = {'서울':'서울특별시', '부산':'부산광역시', '대구':'대구광역시', '인천':'인천광역시', '광주':'광주광역시', '대전':'대전광역시', '울산':'울산광역시', '세종':'세종특별자치시'}

sangMul = pd.read_excel('./data/생활물가지수.xls')
sangMul.fillna('', inplace=True)
sangMul.columns = sangMul.iloc[1]
sangMul_1 = sangMul.drop([0,1])
sangMul_1 = sangMul_1.set_index(sangMul_1['지역'])


sangMul_1 = sangMul_1.apply(lambda x: pd.to_numeric(x, errors='coerce'))
sangMul_1.drop(sangMul_1.columns[0], axis=1, inplace=True)
# sangMul_1.index
sangMul_1.columns.name = None


sangMul_1 = sangMul_1.iloc[1:9]
sangMul_1.rename(city,inplace=True)

sangMul_1 = sangMul_1['2023년 12월'].sort_values()

sangMul_1.plot(kind='barh')
plt.title('생활물가지수 2023년 12월')
plt.xlim(2.5)
```
